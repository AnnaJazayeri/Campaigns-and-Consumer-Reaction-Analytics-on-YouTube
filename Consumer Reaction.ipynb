```python
import os
import json
from googleapiclient.discovery import build

# API key
API_KEY = "AIzaSyB" 

# favoriate topic
favorite_topic = "data science"

# Part 2: 
# getting the information of Video Information
def get_video_information(topic):
    youtube = build("youtube", "v3", developerKey=API_KEY)

    # finding the videos related to the specified topic
    search_response = youtube.search().list(
        q=topic,
        type="video",
        part="id,snippet",
        maxResults=5  # with changing this number the number of video that will retrieve will change
    ).execute()

    videos = []
    for search_result in search_response.get("items", []):
        video_id = search_result["id"]["videoId"]
        video_response = youtube.videos().list(
            id=video_id,
            part="snippet,statistics"
        ).execute()

        # Check if statistics are available for the video
        if "statistics" in video_response["items"][0]:
            video_info = {
                "date_published": video_response["items"][0]["snippet"]["publishedAt"],
                "channel_title": video_response["items"][0]["snippet"]["channelTitle"],
                "title": video_response["items"][0]["snippet"]["title"],
                "description": video_response["items"][0]["snippet"]["description"],
                "statistics": {
                    "view_count": video_response["items"][0]["statistics"].get("viewCount", 0),
                    "like_count": video_response["items"][0]["statistics"].get("likeCount", 0),
                    "dislike_count": video_response["items"][0]["statistics"].get("dislikeCount", 0),
                    "favorite_count": video_response["items"][0]["statistics"].get("favoriteCount", 0),
                    "comment_count": video_response["items"][0]["statistics"].get("commentCount", 0),
                }
            }
            videos.append(video_info)

    return videos

# Part 3:
# Get Statistical Information
def print_video_information(videos):
    for video in videos:
        print(f"Date of Publication: {video['date_published']}")
        print(f"Channel Title: {video['channel_title']}")
        print(f"Title: {video['title']}")
        print(f"Description: {video['description']}")
        print("Statistics:")
        for stat, value in video["statistics"].items():
            print(f"{stat.replace('_', ' ').title()}: {value}")
        print("\n------------------------------------------------------------------------------------------------------------------------------\n")

# Execute the script
videos_data = get_video_information(favorite_topic)
print_video_information(videos_data)

```
